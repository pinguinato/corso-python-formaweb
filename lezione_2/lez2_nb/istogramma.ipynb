```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as ran

N = 1000
x = ran.random_sample(N)
print(x)
print( "x minimo = %le, x massimo = %le " % (x.min(), x.max()) )

```

```python
Nc = 10
pdf, bins = np.histogram( x, Nc, density = True )
print( pdf )
print( bins )
```

```python
nextr = len( bins )
dx = bins[1] - bins[0]
plt.xlabel( r"$x$" )
plt.ylabel( "PDF" )
plt.bar( bins[:nextr-1], pdf, dx, color = "g" )
plt.show()
```
