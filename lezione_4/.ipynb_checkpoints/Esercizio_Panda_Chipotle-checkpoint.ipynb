```python
# l'articolo del datasets più ordinato
most_ordered_items = chipo.groupby('item_name')
most_ordered_items = most_ordered_items.sum()
most_ordered_items = most_ordered_items.sort_values(['quantity'], ascending=False)
most_ordered_items.head(1)
```

```python
# per l'articolo più ordinato, estrarre la quantità ordinata
c = chipo.groupby('item_name')
c = c.sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)
```
